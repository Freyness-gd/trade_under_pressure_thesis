```python
# Import necessary libraries
import geopandas as gpd
import numpy as np
import math


# Function to convert RGB values to hexadecimal color codes
def rgb2hex(r, g, b):
  return "#{:02x}{:02x}{:02x}".format(r, g, b)


# Function to generate a color based on a percentage value
def rg(p):
  d = [255, 0, 0]
  d[1] = int((510 * p) / 100)
  if d[1] > 255:
    d[0] -= d[1] - 255
    d[1] = 255
  return rgb2hex(d[0], d[1], d[2])


# Function to determine the color mapping for a given value
def colourmap(value):
  if value != -99 and not math.isnan(value):
    return rg(value)
  else:
    return "#808080"


# Function to generate random points within valid coordinate ranges
def generate_random_points(df):
  x = np.random.uniform(-180, 180, len(df))
  y = np.random.uniform(-90, 90, len(df))
  return gpd.GeoSeries(gpd.points_from_xy(x, y), crs="EPSG:4326")


# Load the shapefile into a GeoDataFrame
homes = gpd.read_file("../../data/countries/ne_10m_admin_0_countries.shp")

# Add the 'Percentage' column with random integer values between 1 and 100
homes["Percentage"] = np.random.randint(1, 101, len(homes))

# Introduce NaN values in the 'Percentage' column for 20% of the rows
nan_indices = homes.sample(frac=0.2).index
homes.loc[nan_indices, "Percentage"] = np.nan

# Fill NaN values with -99 to indicate missing data
homes["Percentage"] = homes["Percentage"].fillna(-99)

# Add a new geometry column 'func_col' with random points
homes["func_col"] = generate_random_points(homes)

# Set the 'func_col' as the active geometry for visualization
homes = homes.set_geometry("func_col")

# Create the map visualization
m = homes.explore(
  column="Percentage",
  cmap=lambda value: colourmap(value),
  marker_kwds=dict(radius=15, fill=True),
  tooltip=["NAME"],  # Ensure 'NAME' matches the actual column name in the GeoDataFrame
  legend_kwds=dict(caption="Percentage", colorbar=True),
  legend=False,
  name="homes"
)

# Display the map
m
```
